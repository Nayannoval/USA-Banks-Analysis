```python
bac['returns']=((bac['Close']/bac['Close'].shift(1))-1)*100
```

```python
citi['returns']=((citi['Close']/citi['Close'].shift(1))-1)*100
wells['returns']=((wells['Close']/wells['Close'].shift(1))-1)*100
```

```python
wells['returns'].hist(bins=50)
```

```python
plt.figure(figsize=(12,8))
wells['returns'].hist(bins=100,label='Wells Fargo',alpha=0.5)
bac['returns'].hist(bins=100,label='Citi Bank',alpha=0.5)
citi['returns'].hist(bins=100,label='Bank of America',alpha=0.5)
plt.legend()
plt.show()
```

```python
# Making box plots
```

```python
box_df=pd.concat([citi['returns'],wells['returns'],bac['returns']],axis=1)
box_df.columns=['Citi','Wells Fargo','Bank of America']
box_df.plot(kind='box',figsize=(16,6))
```

```python
# Citibank has more volatility on both sides
```

```python
scatter_matrix(box_df,figsize=(8,8),hist_kwds={'bins':50})
plt.show()
```

```python
# Citi and BOA  returns are most correlated as we can see a straight line with more data density
```

```python
# Cumulative return
```

```python
citi['cumulative return']=(1+citi['returns']).cumprod()
bac['cumulative return']=(1+bac['returns']).cumprod()
wells['cumulative return']=(1+wells['returns']).cumprod()
```

```python
citi.head(2)
```

```python
plt.figure(figsize=(16,8))
citi['cumulative return'].plot(label='Citi')
bac['cumulative return'].plot(label='BAC')
wells['cumulative return'].plot(label='Wells Fargo')
```
